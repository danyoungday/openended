```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
baseline_df = pd.read_csv("models/baseline2/logs/progress.csv")
mod_df = pd.read_csv("models/modified/logs/progress.csv")
```

```python
plt.scatter(baseline_df["time/total_timesteps"], baseline_df["rollout/ep_rew_mean"], label="Baseline")
plt.scatter(mod_df["time/total_timesteps"], mod_df["rollout/ep_rew_mean"], label="Modified")
plt.xlabel("Total Timesteps")
plt.ylabel("Mean Episode Reward")

plt.legend()
plt.show()
```
